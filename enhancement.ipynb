```python
#3) import an image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import google.colab.patches as patches
from google.colab.patches import cv2_imshow

# Path to your image
image_path = '/content/drive/MyDrive/Photo.jpg'

# Read the image
image = cv2.imread(image_path)
cv2_imshow(image)
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
image2=np.array([[255, 0, 0, 255],
[0, 0, 0, 0],
[0, 0, 0, 0],
[255, 0, 0, 255]], dtype=np.uint8)
new_size = (800,400)
black_white_image = cv2.resize(image2, new_size) # Resize the original image
# Display the resized image
cv2_imshow(black_white_image)
```

```python
#4) Adjust brightness and contrast of the image
def adjust_brightness_contrast(image, brightness=0, contrast=0):
    beta = brightness
    alpha = contrast / 127.0 + 1.0  # alpha ranges from 1.0 to 3.0
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted

bright_contrast_image = adjust_brightness_contrast(image, brightness=50, contrast=30)
cv2_imshow(bright_contrast_image)

bright_contrast_image2=adjust_brightness_contrast(grey_image, brightness=50, contrast=30)
cv2_imshow(bright_contrast_image2)

black_white_bright_image = adjust_brightness_contrast(black_white_image, brightness=50, contrast=30)
cv2_imshow(black_white_bright_image)
```

```python
#5) find the digital negative of image
negative_image = cv2.bitwise_not(image)
cv2_imshow(negative_image)

negative_image2 = cv2.bitwise_not(grey_image)
cv2_imshow(negative_image2)

negative_image3 = cv2.bitwise_not(black_white_image)
cv2_imshow(negative_image3)
```

```python
#6) Get the red,green,and blue values of each pixel
b, g, r = cv2.split(image)
print(b)
print(g)
print(r)
```

```python
#7) Subtract each color value from 255 and save them as new color values:
b_new = 255 - b
g_new = 255 - g
r_new = 255 - r
```

```python
#8)Create a new pixel value from the modified colors
negative_custom_image = cv2.merge([b_new, g_new, r_new])
```
