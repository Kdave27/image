```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import google.colab.patches as patches
from google.colab.patches import cv2_imshow

# Path to your image
image_path = '/content/drive/MyDrive/Photo.jpg'

image_path_tiff='/content/drive/MyDrive/photojpeg.tiff'


image = cv2.imread(image_path)

image_tiff=cv2.imread(image_path_tiff)
# Display the image
cv2_imshow(image)
cv2_imshow(image_tiff)
```

```python
# Save the image in PNG format (lossless compression)
cv2.imwrite('image_lossless.png', image)

cv2.imwrite('image_tiff_lossless.png',image_tiff)
```

```python
# Save the image in JPEG format (lossy compression)
cv2.imwrite('image_lossy.jpg', image, [int(cv2.IMWRITE_JPEG_QUALITY), -30])

cv2.imwrite('image_tiff_lossy.jpg', image_tiff, [int(cv2.IMWRITE_JPEG_QUALITY), -30])
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load images for plotting
original_image = mpimg.imread(image_path)
lossless_image = mpimg.imread('image_lossless.png')
lossy_image = mpimg.imread('image_lossy.jpg')

original_image2 = mpimg.imread(image_path_tiff)
lossless_image2 = mpimg.imread('image_tiff_lossless.png')
lossy_image2 = mpimg.imread('image_tiff_lossy.jpg')
# Plot images
plt.figure(figsize=(15, 5))

plt.subplot(1, 6, 1)
plt.title('Original Image')
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 6, 2)
plt.title('Lossless Compression (PNG)')
plt.imshow(lossless_image)
plt.axis('off')

plt.subplot(1, 6, 3)
plt.title('Lossy Compression (JPEG)')
plt.imshow(lossy_image)
plt.axis('off')

plt.subplot(1, 6, 4)
plt.title('Original Image')
plt.imshow(original_image2)
plt.axis('off')

plt.subplot(1, 6, 5)
plt.title('Lossless Compression (PNG)')
plt.imshow(lossless_image2)
plt.axis('off')

plt.subplot(1, 6, 6)
plt.title('Lossy Compression (JPEG)')
plt.imshow(lossy_image2)
plt.axis('off')

plt.show()
```
