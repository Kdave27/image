```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import google.colab.patches as patches
from google.colab.patches import cv2_imshow

# Path to your image
image_path = '/content/drive/MyDrive/Photo.jpg'

image = cv2.imread(image_path)
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Color Image')
plt.subplot(1, 2, 2)
plt.imshow(grey_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
#4 detect the boundaries in the image
edges = cv2.Canny(image, 100, 200)

# Display the edges
plt.figure(figsize=(10,5))
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.show()
edges1 = cv2.Canny(grey_image, 100, 200)

# Display the edges
plt.figure(figsize=(10,5))
plt.imshow(edges1, cmap='gray')
plt.title('Edges')
plt.show()
```

```python

dct1= cv2.dct(np.float32(grey_image))

plt.figure(figsize=(10,5))
plt.imshow(dct1, cmap='gray')
plt.title('DCT')
plt.show

```

```python
#use detectMultiScale function to detect an object
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(grey_image, scaleFactor=1.1, minNeighbors=5,minSize=(30,30))
# Draw rectangles around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the result
plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.show()

for (x, y, w, h) in faces:
    cv2.rectangle(grey_image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the result
plt.figure(figsize=(10,5))
plt.imshow(cv2.cvtColor(grey_image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.show()
```
