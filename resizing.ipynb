```python
#4) read image into notebook using openCv
import cv2

image_path='/content/drive/MyDrive/Photo.jpg'

image=cv2.imread(image_path)
```

```python
#5) Display the image
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
# Check the size (shape) and type of the image
print("Shape of the image:", image.shape)
print("Type of the image:", type(image))

new_size = (800,400)
resized_image = cv2.resize(image, new_size)
# Display the resized image
cv2_imshow(resized_image)

copied_image=resized_image.copy()
cv2_imshow(copied_image)
```

```python
#6) Write the displayed image into new file
new_image_path='/content/drive/MyDrive/photo5.jpg'
cv2.imwrite(new_image_path, resized_image)
```
